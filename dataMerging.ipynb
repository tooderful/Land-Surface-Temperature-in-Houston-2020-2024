```python
import pandas as pd
df = pd.read_csv('harrisCountyProcessedData/forestSpatialConfigurations.csv')
```

```python
harrisForestSpatialConfig_df = df.iloc[:, [3, 4, 8, 15, 16]]
harrisForestSpatialConfig_df.shape
```

```python
fctImpDf = pd.read_csv('harrisCountyProcessedData/harrisFctImp.csv')
fctImpDf.shape
```

```python
result_inner = pd.merge(harrisForestSpatialConfig_df, fctImpDf, on='CTBKEY')
result_inner.drop(['ZONE_CODE', 'COUNT', 'AREA'], axis=1, inplace=True)
result_inner.rename(columns={'MEAN':'FctImp'}, inplace=True)
```

```python
lstDf = pd.read_csv('harrisCountyProcessedData/harrisLST.csv')
lstDf.shape
```

```python
harris_df = pd.merge(result_inner, lstDf, on='CTBKEY')
harris_df.drop(['ZONE_CODE', 'COUNT', 'AREA'], axis=1, inplace=True)
harris_df.rename(columns={'MEAN':'LST'}, inplace=True)
harris_df.shape
```

```python
ndviDf = pd.read_csv('harrisCountyProcessedData/harrisNDVI.csv')
ndviDf.shape
```

```python
harris_df = pd.merge(harris_df, ndviDf, on='CTBKEY')
harris_df.drop(['ZONE_CODE', 'COUNT', 'AREA'], axis=1, inplace=True)
harris_df.rename(columns={'MEAN':'NDVI'}, inplace=True)
harris_df.shape
```

```python
perCoverForestDf = pd.read_csv('harrisCountyProcessedData/perCoverForest.csv')
perCoverForestDf.shape
```

```python
harrisForest_df = pd.merge(harris_df, perCoverForestDf, on='CTBKEY')
harrisForest_df.drop(['VALUE_0', 'VALUE_1', 'Total_Cells'], axis=1, inplace=True)
harrisForest_df.to_csv('harrisDataF.csv', index=False)
```
